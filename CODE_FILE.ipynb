```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\POOJA\\Downloads\\SMSSpamCollection.csv', delimiter='\t', header=None)
print(df.shape)
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df[0]=df[0].replace(['ham','spam'],['Not Spam','Spam'])
df = df.rename(columns={0:'Category',1:'Mail'})
```

```python
df.head()
```

```python
mail=df['Mail']
cat=df['Category']
```

```python
from sklearn.model_selection import train_test_split
mail_train, mail_test, cat_train, cat_test = train_test_split(
    mail, cat, test_size=0.33, random_state=10)
print(cat_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
 # convert text to numerical data
cv=CountVectorizer(stop_words='english')
features = cv.fit_transform(mail_train)
print(features)
```

```python
# creating model
from sklearn.naive_bayes import MultinomialNB
model= MultinomialNB()
model.fit(features , cat_train)
```

```python
#testing 
features_test=cv.transform(mail_test)
category_pred=model.predict(features_test)
category_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(category_pred,cat_test)
print(score)
```

```python
print(classification_report(category_pred,cat_test))
```
